```
import pandas as pd
dataframe=pd.read_csv('data.csv', index_col=0)
dataframe
```

```
#Arrangement des données par ordre croissant
dataframe = dataframe[['2009', '2011', '2015', '2016', '2017', '2018']]
dataframe
```

```
#On fait un graghique montrant l'évolution des coûts de scolaritéé pour les écoles de commerces entre 2009 et 2018
dataframe.transpose().plot(title='Evolution des couts de scolaritee pour les ecoles de commerces entre 2009 et 2018', figsize=(20,10));
```

```
#On fait un graghique montrant le coût moyen et médian des écoles de commerce françaises par année
mean = dataframe.mean()
mean.plot(title='Cout moyen et median des ecoles de commerce Francaises par année:')
```

```
if mean['2009'] < mean['2018']:
    print('Le cout moyen a augmente entre 2009 et 2018')
else:
    print('Le cout moyen a baisse entre 2009 et 2018')
```

```
median = dataframe.median()
if mean['2009'] < mean['2018']:
    print('Le cout median a augmente entre 2009 et 2018')
else:
    print('Le cout median a baisse entre 2009 et 2018')
```

```
pourcentages = dataframe.transpose().pct_change()
# Affichage en pourcentages
pourcentages.style.format("{:.2%}")
```

```
annee_max = pourcentages.transpose().max().idxmax()
print('La hausse la plus importante de coût de scolarité a eu lien en', annee_max)
```
